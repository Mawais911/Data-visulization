```python
# Data
df = sns.load_dataset("titanic")
```

```python
# check the data types and column information
df.info()
```

```python
# scatter plot
sns.scatterplot(data=df, x='age', y='fare')
```

```python
# grouping on the base of sex (categorical variable)
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# figure size
plt.figure(figsize=(10, 6))
# scatter plot with figure size
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# Change the scatter point color
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

```python
# scatter plot
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')

# saving the plot 300dpi, 600dpi, png, pdf
plt.savefig('../report/scatter_plot_age_vs_fare.png', dpi=300)
```

```python
# change the font and font scale
sns.set_theme(style="darkgrid", font_scale=1 )
# scatter plot
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

```python
# change the font and font scale
sns.set_theme(style="darkgrid", font_scale=1 )
# scatter plot
plt.figure(figsize=(16, 6))
palette = {'male': 'green', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette, size='age')
# add labels
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter plot of Age vs Fare')
```

# **Box Plot**

```python
# change style of the plot
sns.set_style("whitegrid")
# Box plot
palette = {'male': 'green', 'female': 'red'}
sns.boxplot(data=df, x='sex', y='age', hue='sex', palette=palette)
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Box plot of Sex vs Age')
```

> change the hue

```python
# change style of the plot
sns.set_style("whitegrid")
# Box plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Box plot of Sex vs Age')
```

> Change legend location and title

```python
# change style of the plot
sns.set_style("whitegrid")
# Box plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Box plot of Sex vs Age')
# change legend location and title
plt.legend(title='Class Name', loc='upper right')
```

# **Boxen Plot**

```python
# change style of the plot
sns.set_style("whitegrid")
# Boxen plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Class Name and fare')
# change legend location and title
plt.legend(title='Class Name', loc='upper right')
```

> Change the x-tiks lables and Rotations

```python
# change style of the plot
sns.set_style("whitegrid")
# Boxen plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Boxen plot of Age, based on Class Name and fare')
# change legend location and title
plt.legend(title='Class Name', loc='upper right')
# change the x-tiks lables
plt.xticks(rotation=45, ha='right',ticks=[0,1], labels=['Male', 'Female'])
```

# **Bar plot**

```python
# change style of the plot
sns.set_style("whitegrid")
# Bar plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette)
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Class Name and fare')
# change legend location and title
plt.legend(title='Class Name', loc='upper right')
# change the x-tiks lables
plt.xticks(rotation=45, ha='right',ticks=[0,1], labels=['Male', 'Female'])
```

> add error bar cap size

```python
# change style of the plot
sns.set_style("whitegrid")
# Bar plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.barplot(data=df, x='sex', y='age', hue='class', palette=palette, capsize=0.1) # adding capsize
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Bar plot of Age, based on Class Name and fare')
# change legend location and title
plt.legend(title='Class Name', loc='upper right')
# change the x-tiks lables
plt.xticks(rotation=45, ha='right',ticks=[0,1], labels=['Male', 'Female'])
```

# **voilen Plot**

```python
# change style of the plot
sns.set_style("whitegrid")
# violin plot
plt.figure(figsize=(13, 6))
palette = {'First': 'green', 'Second': 'red', 'Third': 'blue'}
sns.violinplot(data=df, x='sex', y='age', hue='class', palette=palette) 
# add labels
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title('Violin plot of Age, based on Class Name and fare')
# change legend location and title
plt.legend(title='Class Name', loc='upper right')
# change the x-tiks lables
plt.xticks(rotation=45, ha='right',ticks=[0,1], labels=['Male', 'Female'])
```
